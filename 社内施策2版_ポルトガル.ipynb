```python
llrt <- function(mfull,m){
  loglik_full <- sum(mfull$y * log(mfull$fitted.values) + (1 - mfull$y) * log(1-mfull$fitted.values))
  logLik_dvar <- sum(m$y * log(m$fitted.values) + (1 - m$y) * log(1-m$fitted.values))
  chisq <- -2 * (logLik_dvar  - loglik_full)
  df <- summary(mfull)$df[1] - summary(m)$df[1]
  p <- 1- pchisq(chisq,df)
  round(c(chisq , df ,p ),3)
  }
```

```python

# カテゴリ変数作成

samp$previous_c <- samp$previous
samp$pdays_c <- samp$pdays

samp[samp$poutcome == 'unknown',]$previous_c <- 0
samp[samp$poutcome == 'unknown',]$pdays_c <- 0

#カテゴリ化
#samp$C_age <- as.factor(samp$age)
samp$C_age <- Recode(samp$age, '0:29 = 29 ; 30:34 =34 ; 34:39 = 39 ; 40:44 = 44 ; 45:49 = 49 ; 50:54 = 54 ; 55:59 = 59 ;  else = 60' )
samp$C_age <- as.factor(samp$C_age)
samp$C_job <- as.factor(samp$job)
samp$C_marital <- as.factor(samp$marital)
samp$C_education <- as.factor(samp$education)
samp$C_default <- as.factor(samp$default)
samp$C_balance <- as.factor(samp$balance)
samp$C_housing <- as.factor(samp$housing)
samp$C_loan<- as.factor(samp$loan)
#samp$C_contact <- as.factor(samp$contact)
#samp$C_day <- as.factor(samp$day)
#samp$C_month <- as.factor(samp$month)
#samp$C_duration <- as.factor(samp$duration)
#samp$C_campaign <- as.factor(samp$campaign)
#samp$C_pdays <- as.factor(samp$pdays)
samp$C_pdays <- Recode(samp$pdays_c,'-1 = -1 ; 0:60 = 60 ; 61:90 = 90 ; 91:120 = 120 ;  else = 999')
samp$C_pdays <- as.factor(samp$C_pdays)
samp$C_pdays_b <- Recode(samp$pdays,'-1 = -1 ; 0:30 = 30 ; 31:60 = 60 ; 61:90 = 90 ; 91:120 = 120 ;  else = 999')
samp$C_pdays_b <- as.factor(samp$C_pdays_b)
#samp$C_previous <- as.factor(samp$previous)
samp$C_previous <- Recode(samp$previous_c,'0 = 0 ; 1 = 1 ; 2 = 2 ;  else = 3')
samp$C_previous <- as.factor(samp$C_previous)
samp$C_previous_b <- Recode(samp$previous,'0 = 0 ; 1 = 1 ; 2 = 2 ;  else = 3')
samp$C_previous_b <- as.factor(samp$C_previous_b)
samp$C_poutcome <- as.factor(samp$poutcome)
samp$C_y <- as.factor(samp$y)

#目的変数をフラグに
samp$yF <- if_else(samp$y == "yes",1,0)
samp$C_yF <- as.factor(samp$yF)



# カテゴリ変数　まとめ
##職業（無職、不明は同カテゴリ
# samp$C_job_02 <- Recode(samp$C_job, ' 12=11 ;  4=3 ; 8=3')

#最終連絡月　3・9・12月 / それ以外　の２区分(基礎集計グラフから)
# samp$C_month_02 <- Recode(samp$C_month,'3 = 1 ;9 = 1 ; 12 = 1 ; else=0')

#最終連絡日付　1・10・20・30・31日 / それ以外　の２区分(基礎集計グラフから)
# samp$C_day_02 <- Recode(samp$C_day,'1 = 1 ; 10 = 1 ; 20 = 1 ; 30 = 1 ; 31 = 1 ; else=0')



#連続値変数作成

#各変数　上限設定

##年齢
samp$age_60 <- pmin(samp$age,60)

#＃平均残高
samp$balance_5000 <- pmax(samp$balance,-1)
samp$balance_5000 <- pmin(samp$balance_5000,5000)

##現在のマーケティングキャンペーンにおける顧客への連絡回数
#samp$campaign_11 <- pmin(samp$campaign,11)

##以前のマーケティングキャンペーンにおける顧客への最終連絡日からの経過日数
#samp$F_pdays <- 0
#samp[samp$pdays == -1,]$F_pdays <- 1
#samp$F_pdays <- as.factor(samp$F_pdays)

##以前のマーケティングキャンペーンにおける顧客への連絡回数
#samp$previous_11 <- pmin(samp$previous,11)


#平滑化
library(splines)
i <- 3

B_age <- bs(samp$age_60,df=i)
B_balance <- bs(samp$balance_5000,df = i)
#B_campaign <- bs(samp$campaign_11,df = i)
#B_previous <- bs(samp$previous_11,df = i)

samp[,"B_age"] <- B_age
samp[,"B_balance"] <- B_balance
#samp[,"B_campaign"] <- B_campaign
#samp[,"B_previous"] <- B_previous

#５回目での変更部分（変数微調整）

##以前のマーケティングキャンペーンにおける顧客への最終連絡日からの経過日数
#samp$pdays_120 <- pmin(samp$pdays,120)

##以前のマーケティングキャンペーンの結果
# samp$C_poutcome <- Recode(samp$C_poutcome, ' 9=8')

#B_pdays <- bs(samp$pdays_120,df = i)

#samp[,"B_pdays"] <- B_pdays

#summary(samp)
```

```python
#サンプリング
set.seed(3)
samp50 <- samp %>% sample_frac(0.5)


#全部の組み合わせの変数作成
crossstring <-"B_age + C_job + C_marital+ C_education +  C_default + B_balance + C_housing + C_loan + C_pdays_b + C_previous_b + C_poutcome"

w <-"B_age : C_marital+ B_age : C_job+ B_age : C_education +  C_age : B_balance + B_age : C_housing + B_age : C_loan + B_age : C_pdays + B_age : C_previous + B_age : C_poutcome"
crossstring <- str_flatten(c(w,crossstring)," + ")

w <- "C_default  : C_loan + C_default : C_housing + C_education : C_default"
crossstring <- str_flatten(c(w,crossstring)," + ")

w <- "C_job : C_marital + C_job : B_balance + C_job : C_housing + C_job : C_loan + C_job : C_pdays + C_job : C_previous + C_job : C_poutcome"
crossstring <- str_flatten(c(w,crossstring)," + ")

w <-"C_marital:C_pdays + C_education:C_pdays + B_balance:C_pdays + C_housing:C_pdays + C_loan:C_pdays "
crossstring <- str_flatten(c(w,crossstring)," + ")


for (i in 1:7) {
  for (j in 1:7) {
    w <- str_flatten(c(switch(i,               # 指定された条件
         "1"  = ("C_marital"),
         "2"  = ("C_education"),
         "3" = ("B_balance"),
         "4" = ("C_housing"),
         "5" = ("C_loan"),
         "6" = ("C_previous"),
         "7" = ("C_poutcome"))
,
        switch(j,               # 指定された条件
         "1"  = ("C_marital"),
         "2"  = ("C_education"),
         "3" = ("B_balance"),
         "4" = ("C_housing"),
         "5" = ("C_loan"),
         "6" = ("C_previous"),
         "7" = ("C_poutcome")))," : ")

      if(i>=j){
      }
      else{
      crossstring <- str_flatten(c(w,crossstring)," + ")
      }
     }
}

crossstring
```

```python
# ニューラルネットワークによる予測
modelNnet_ct <- train(
  as.formula(paste("C_yF ~ ",crossstring, sep="")),
  data = samp50,
  method = "nnet",
  tuneGrid = expand.grid(size=c(1:3), decay=seq(0.1, 1, 0.1)),
  linout = FALSE
)
```

```python
caret::varImp(modelNnet_ct, scale = TRUE)
```

```python
modelNnet_in_result_data <- data.frame(samp50,p=predict(modelNnet_ct,samp50,type="prob"))
summary(modelNnet_in_result_data)
```

```python
#サンプリング
set.seed(3)
samp50 <- samp %>% sample_frac(0.5)


#全部の組み合わせの変数作成
crossstring <-"C_previous1 : C_poutcome1 + C_loan1 : C_poutcome2 + C_loan2 : C_previous2 + C_housing1 : C_poutcome3 + C_housing2 : C_previous3+ C_housing3 : C_loan3 + B_balanceA : C_poutcome4 + B_balanceB : C_previous4 + B_balanceC : C_loan4 + B_balanceD : C_housing4 + C_education1 : C_poutcome5 + C_education2 : C_previous5 + C_education3 : C_loan5 + C_education4 : C_housing5 + C_education5 : B_balanceE + C_marital1 : C_poutcome6 + C_marital2 : C_previous6 + C_marital3 : C_loan6 + C_marital4 : C_housing6 + C_marital5 : B_balanceF + C_marital6 : C_education6 + C_marital7:C_pdays1 + C_education7:C_pdays2 + B_balanceG:C_pdays3 + C_housing7:C_pdays4 + C_loan7:C_pdays5 + C_job1 : C_marital8 + C_job2 : B_balanceH + C_job3 : C_housing8 + C_job4 : C_loan8 + C_job5 : C_pdays6 + C_job6 : C_previous7 + C_job7 : C_poutcome7 + C_default1  : C_loan9 + C_default2 : C_housing9 + C_education8 : C_default3 + B_ageA : C_marital9 + B_ageB : C_job8 + B_ageC : C_education9 + C_age : B_balanceI + B_ageE : C_housing10 + B_ageF : C_loan10 + B_ageG : C_pdays7 + B_ageH : C_previous8 + B_ageI : C_poutcome8 + B_age + C_job + C_marital + C_education +  C_default + B_balance + C_housing + C_loan + C_pdays_b + C_previous_b + C_poutcome"

crossstring
```

```python
# ニューラルネットワークによる予測
modelNnet_ct <- train(
  as.formula(paste("C_yF ~ ",crossstring, sep="")),
  data = samp50,
  method = "nnet",
  tuneGrid = expand.grid(size=c(1:3), decay=seq(0.1, 1, 0.1)),
  linout = FALSE
)
```

```python
caret::varImp(modelNnet_ct, scale = TRUE)
```

```python
modelNnet_in_result_data <- data.frame(samp50,p=predict(modelNnet_ct,samp50,type="prob"))
#modelNnet_in_result_data <- data.frame(samp50,p=predict(modelNnet,samp50))

write.table(modelNnet_in_result_data, file = "/content/modelNnet_in_result_data_pi.csv",row.names=F,sep=",")
#saveRDS(modelNnet, file = "/content/modelNnet.obj")
#saveRDS(modelNnet_ct, file = "/content/modelNnet.obj")

d_arcap <- modelNnet_in_result_data

#目的変数名
targetName <- "yF"
targetPred <- d_arcap[,targetName]

#ソートする変数、予測確率、GNO、CR
sortName <- "p.1"
sortPred <- d_arcap[,sortName]

#ARの計算
pred <- prediction(sortPred,targetPred)
AUC <- as.numeric( performance(pred,"auc")@y.values )
AR <- 2 *(AUC-0.5) * 100 #AR値

#CAP値の計算
#予測確率（GNO、CR）を照準に件数等分１０分類
d_arcap$CP10 <- bin.var( sortPred, bins=10, method='proportions',labels =FALSE)

#区分別のデフォルト件数
dcount <- tapply(targetPred,list(CP10=d_arcap$CP10),sum,na.rm=TRUE)

#CAP20値
CAP20 <- sum(dcount[9:10]) / sum(dcount) * 100

#CAP30値
CAP30 <- sum(dcount[8:10]) / sum(dcount) * 100

#CAP70値
CAP70 <- sum(dcount[4:10]) / sum(dcount) * 100

#結果
round(c(AR,CAP20,CAP30,CAP70),2)
nrow(d_arcap)
```

```python
#サンプリング
set.seed(3)
samp50 <- samp %>% sample_frac(0.5)


#全部の組み合わせの変数作成
crossstring <-"B_age + C_job + C_marital + C_education + C_default + B_balance + C_housing + C_loan + C_pdays_b + C_poutcome + C_loan:C_poutcome + C_housing:C_loan + B_balance:C_poutcome + C_education:C_loan +  C_education:C_housing + C_education:B_balance + C_marital:C_previous +  C_housing:C_pdays + C_loan:C_pdays + C_default:C_loan + C_default:C_housing +  B_balance:C_age + B_age:C_housing + B_age:C_loan + B_age:C_poutcome"
crossstring
```

```python
modelRF_sw  <- train(
  as.formula(paste("C_yF ~ ",crossstring, sep="")),
  data = samp50,
  method = "rf",
  tuneLength = 2,
  )
```

```python
caret::varImp(modelRF_sw, scale = TRUE)
```

```python
#modelXgb_in_result_data <- data.frame(samp50,p=predict(modelXgboostTree,samp50))
modelRF_sw_in_result_data <- data.frame(samp50,p=predict(modelRF_sw,samp50,type="prob"))
write.table(modelRF_sw_in_result_data, file = "/content/modelRF_sw_in_result_data.csv",row.names=F,sep=",")
#saveRDS(modelXgboostTree, file = "/content/modelXgboostTree.obj")
saveRDS(modelRF_sw, file = "/content/modelRF_sw.obj")

d_arcap <- modelRF_sw_in_result_data

#目的変数名
targetName <- "yF"
targetPred <- d_arcap[,targetName]

#ソートする変数、予測確率、GNO、CR
sortName <- "p.1"
sortPred <- d_arcap[,sortName]

#ARの計算
pred <- prediction(sortPred,targetPred)
AUC <- as.numeric( performance(pred,"auc")@y.values )
AR <- 2 *(AUC-0.5) * 100 #AR値

#CAP値の計算
#予測確率（GNO、CR）を照準に件数等分１０分類
d_arcap$CP10 <- bin.var( sortPred, bins=10, method='proportions',labels =FALSE)

#区分別のデフォルト件数
dcount <- tapply(targetPred,list(CP10=d_arcap$CP10),sum,na.rm=TRUE)

#CAP20値
CAP20 <- sum(dcount[9:10]) / sum(dcount) * 100

#CAP30値
CAP30 <- sum(dcount[8:10]) / sum(dcount) * 100

#CAP70値
CAP70 <- sum(dcount[4:10]) / sum(dcount) * 100

#結果
round(c(AR,CAP20,CAP30,CAP70),2)
nrow(d_arcap)
```

```python

# カテゴリ変数作成

samp$previous_c <- samp$previous
samp$pdays_c <- samp$pdays

samp[samp$poutcome == 'unknown',]$previous_c <- 0
samp[samp$poutcome == 'unknown',]$pdays_c <- 0

#カテゴリ化
#samp$C_age <- as.factor(samp$age)
samp$C_age <- Recode(samp$age, '0:29 = 29 ; 30:34 =34 ; 34:39 = 39 ; 40:44 = 44 ; 45:49 = 49 ; 50:54 = 54 ; 55:59 = 59 ;  else = 60' )
samp$C_age <- as.factor(samp$C_age)
samp$C_job <- as.factor(samp$job)
samp$C_marital <- as.factor(samp$marital)
samp$C_education <- as.factor(samp$education)
samp$C_default <- as.factor(samp$default)
samp$C_balance <- as.factor(samp$balance)
samp$C_housing <- as.factor(samp$housing)
samp$C_loan<- as.factor(samp$loan)
#samp$C_contact <- as.factor(samp$contact)
#samp$C_day <- as.factor(samp$day)
#samp$C_month <- as.factor(samp$month)
#samp$C_duration <- as.factor(samp$duration)
#samp$C_campaign <- as.factor(samp$campaign)
#samp$C_pdays <- as.factor(samp$pdays)
samp$C_pdays <- Recode(samp$pdays_c,'-1 = -1 ; 0:60 = 60 ; 61:90 = 90 ; 91:120 = 120 ;  else = 999')
samp$C_pdays <- as.factor(samp$C_pdays)
samp$C_pdays_b <- Recode(samp$pdays,'-1 = -1 ; 0:30 = 30 ; 31:60 = 60 ; 61:90 = 90 ; 91:120 = 120 ;  else = 999')
samp$C_pdays_b <- as.factor(samp$C_pdays_b)
#samp$C_previous <- as.factor(samp$previous)
samp$C_previous <- Recode(samp$previous_c,'0 = 0 ; 1 = 1 ; 2 = 2 ;  else = 3')
samp$C_previous <- as.factor(samp$C_previous)
samp$C_previous_b <- Recode(samp$previous,'0 = 0 ; 1 = 1 ; 2 = 2 ;  else = 3')
samp$C_previous_b <- as.factor(samp$C_previous_b)
samp$C_poutcome <- as.factor(samp$poutcome)
samp$C_y <- as.factor(samp$y)

#目的変数をフラグに
samp$yF <- if_else(samp$y == "yes",1,0)
samp$C_yF <- as.factor(samp$yF)



# カテゴリ変数　まとめ
##職業（無職、不明は同カテゴリ
# samp$C_job_02 <- Recode(samp$C_job, ' 12=11 ;  4=3 ; 8=3')

#最終連絡月　3・9・12月 / それ以外　の２区分(基礎集計グラフから)
# samp$C_month_02 <- Recode(samp$C_month,'3 = 1 ;9 = 1 ; 12 = 1 ; else=0')

#最終連絡日付　1・10・20・30・31日 / それ以外　の２区分(基礎集計グラフから)
# samp$C_day_02 <- Recode(samp$C_day,'1 = 1 ; 10 = 1 ; 20 = 1 ; 30 = 1 ; 31 = 1 ; else=0')



#連続値変数作成

#各変数　上限設定

##年齢
samp$age_60 <- pmin(samp$age,60)

#＃平均残高
samp$balance_5000 <- pmax(samp$balance,-1)
samp$balance_5000 <- pmin(samp$balance_5000,5000)

##現在のマーケティングキャンペーンにおける顧客への連絡回数
#samp$campaign_11 <- pmin(samp$campaign,11)

##以前のマーケティングキャンペーンにおける顧客への最終連絡日からの経過日数
#samp$F_pdays <- 0
#samp[samp$pdays == -1,]$F_pdays <- 1
#samp$F_pdays <- as.factor(samp$F_pdays)

##以前のマーケティングキャンペーンにおける顧客への連絡回数
#samp$previous_11 <- pmin(samp$previous,11)


#平滑化
library(splines)
i <- 3

B_age <- bs(samp$age_60,df=i)
B_balance <- bs(samp$balance_5000,df = i)
#B_campaign <- bs(samp$campaign_11,df = i)
#B_previous <- bs(samp$previous_11,df = i)

samp[,"B_age"] <- B_age
samp[,"B_balance"] <- B_balance
#samp[,"B_campaign"] <- B_campaign
#samp[,"B_previous"] <- B_previous

#５回目での変更部分（変数微調整）

##以前のマーケティングキャンペーンにおける顧客への最終連絡日からの経過日数
#samp$pdays_120 <- pmin(samp$pdays,120)

##以前のマーケティングキャンペーンの結果
# samp$C_poutcome <- Recode(samp$C_poutcome, ' 9=8')

#B_pdays <- bs(samp$pdays_120,df = i)

#samp[,"B_pdays"] <- B_pdays

#summary(samp)
```

```python
#サンプリング
set.seed(3)
samp50 <- samp %>% sample_frac(0.5)


#全部の組み合わせの変数作成
crossstring <-"B_age + C_marital + C_education + C_default + B_balance +  C_housing + C_loan + C_pdays_b + C_poutcome + C_housing:C_pdays + C_housing:C_loan + B_balance:C_poutcome + C_marital:C_previous +  C_default:C_housing + B_balance:C_age + B_age:C_housing +   B_age:C_poutcome"
crossstring
```

```python
crossstring
```

```python
modelSVM <- train(
  as.formula(paste("C_y ~ ",crossstring, sep="")),
  data = samp50,
　trControl = trainControl(classProbs =  TRUE),
  method = "svmRadial"
)
```

```python
caret::varImp(modelSVM, scale = TRUE)
```

```python
modelSVM_in_result_data <- data.frame(samp50,predict(modelSVM,samp50,type="prob"))
write.table(modelSVM_in_result_data, file = "/content/modelSVM_in_result_data.csv",row.names=F,sep=",")
saveRDS(modelSVM, file = "/content/modelSVM.obj")

#目的変数名
targetName <- "C_y"
targetPred <- modelSVM_in_result_data[,targetName]

#ソートする変数、予測確率、GNO、CR
sortName <- "yes"
sortPred <- modelSVM_out_result_data[,sortName]

#ARの計算
pred <- prediction(sortPred,targetPred)
AUC <- as.numeric( performance(pred,"auc")@y.values )
AR <- 2 *(AUC-0.5) * 100 #AR値

AR
```
